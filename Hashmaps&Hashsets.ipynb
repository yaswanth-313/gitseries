```python
#hashsets
s= set()
print(s)
```

```python
#addig to set -O(1)
s.add(1)
s.add(2)
s.add(3)

print(s)
```

```python
#lookup for element - O(1)
if 2 in s:
  print (True)
```

```python
#deleting the element - O(1)

s.remove(3)
print(s)
```

```python
string="aaabababcccdddde"
sh= set(string)
print(sh)
```

```python
#loop ove i a set - O(n)

for x in s:
  print(s)
```

```python
#hashmaps - Dictionaries
d={'greg':1 , 'steve':2 , 'rob':3}
print(d)
```

```python
#addingg the elments in dict - O(1)
d['laha']=4
print(d)
```

```python
#check for presence in dict - O(1)
if 'laha' in d:
  print(True)
```

```python
print(d['laha'])
```

```python
for key, val in d.items():
  print(f'key {key}: val {val}')
```

```python
#countthe frequency of a string using a dict
kl = "pythonprogramming"
frequency = {}
for char in kl:
  if char in frequency:
    frequency[char] += 1
  else:
    frequency[char] = 1
print(frequency)
```
